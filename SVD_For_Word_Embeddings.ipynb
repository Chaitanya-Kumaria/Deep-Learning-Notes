```python

import numpy as np
import pandas as pd
np.set_printoptions(linewidth=120)
```

```python
# Corpus
sentences = ['Human machine interface for computer applications','User opinion of computer system response time','User interface management system','System engineering for improved response time']
```

```python
# Vocabulary
vocab=list(set((" ".join(sentences)).split()))
vocab
```

```python
text_data=[]
for i in sentences:
    text_data.append(i.split())
text_data
```

```python
window_size = 1
w=[]
for i in sentences:
    i=i.split(' ') #if i is list
    for k in range(len(i)-window_size+1):
        for l in range(k+1,k+window_size+1):
            if l<=len(i)-1:
                w.append([i[k],i[l]])
w
```

```python
w1=[x[::-1] for x in w]
w.extend(w1)
```

```python
a=np.zeros((len(vocab),len(vocab)))
df=pd.DataFrame(a,index=vocab,columns=vocab)
for i in w:
    df.at[i[0],i[1]]+=1
```

```python
# Co occurence matrix
df
```

```python
from numpy import array
from scipy.linalg import svd
from numpy import diag
from numpy import dot
# define a matrix
A = array(df)
print('Matrix A is: \n')
print(A)
# SVD
U, s, VT = svd(A)
print('*'*120)
print('Matrix U is: \n')
print(U)
print('*'*120)
Sigma = diag(s)
print('Matrix Sigma is: \n')
print(Sigma)
print('*'*120)
print('Matrix VT is: \n')
print(VT)
```

```python
U.shape,Sigma.shape,VT.shape
```

```python
k = 3
U=U[:,:k]
Sigma=Sigma[:k,:k]
VT=VT[:k,:]
U.shape,Sigma.shape,VT.shape
```

```python
U_df=pd.DataFrame(U,index=vocab)
U_df
```
